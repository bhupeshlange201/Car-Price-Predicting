```python
df['Car_Brand']=df['CarName'].apply(lambda x: x.split(' ')[0])
```

```python
df['Car_Brand'].value_counts()
```

```python
df['Car_Brand']= df['Car_Brand'].str.replace('toyouta','toyota')
df['Car_Brand']= df['Car_Brand'].str.replace('porcshce','porsche')
df['Car_Brand']= df['Car_Brand'].str.replace('Nissan','nissan')
df['Car_Brand']= df['Car_Brand'].str.replace('maxda','mazda')
df['Car_Brand']= df['Car_Brand'].str.replace('vw','volkswagen')
df['Car_Brand']= df['Car_Brand'].str.replace('vokswagen','volkswagen')
```

```python
df['Car_Brand'].value_counts()
```

```python
df.columns
```

```python
df['doornumber']= df['doornumber'].replace('four',4)
df['doornumber']= df['doornumber'].replace('two',2)
```

```python
df['cylindernumber']=df['cylindernumber'].map({'four': 4,'six':6,'five':5,'eight':8,'two':2,'twelve':12,'three':3})
```

```python
df['cylindernumber'].value_counts()
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop('CarName',axis=1,inplace=True)
```

```python
df.shape
```

```python
dummy= pd.get_dummies(df,drop_first=True)
```

```python
dummy.shape
```

```python
dummy.head()
```

```python
x= dummy.drop('price',axis=1)
y= dummy['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
var=['symboling', 'doornumber', 'wheelbase', 'carlength', 'carwidth',
       'carheight', 'curbweight', 'cylindernumber', 'enginesize', 'boreratio','stroke',
       'compressionratio', 'horsepower', 'peakrpm','citympg','highwaympg']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train[var]=scaler.fit_transform(x_train[var])
```

```python
x_train.shape
```

```python
x_test[var]=scaler.transform(x_test[var])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
features_columns=x_train_59.drop('const',axis=1)
vif=pd.DataFrame()
vif['features']=features_columns.columns
vif['VIF']=[variance_inflation_factor(features_columns.values,i) for i in range(features_columns.shape[1])]
vif['VIF']=round(vif['VIF'],2)
vif=vif.sort_values(by="VIF",ascending=False)
vif
```
